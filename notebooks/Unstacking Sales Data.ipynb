```python
import pandas as pd
brough = pd.read_csv('brughastacked_r.csv')
brough.head()
```

```python
# Unstack and assign column names to Posttest and Pretest
brough_unstack = (brough
       .set_index(['Store Number', 'Panel', 'Period'])
       .unstack())
brough_unstack.columns = brough_unstack.columns.droplevel([0])
brough_unstack.rename_axis([None], axis=1).reset_index()
```

```python
# Save file as Brougha Unstacked
brough_unstack.to_csv('broughaunstacked_r.csv')

# Note, you can save the file with the columns reordered here:
# brough_unstack[['Store Number', 'Panel', 'Pretest', 'Posttest']].to_csv('broughaunstacked_r.csv')
```
