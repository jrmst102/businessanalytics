```python
# Read Smoke Detectors file

import pandas as pd
df = pd.read_csv('smokedetectors_r.csv')
print(df.head())
```

```python
# Calculate correlation
correlation = df['Homes With'].corr(df['Deaths per Mil'])
print("Correlation between Homes with Smoke Detectors and Deaths per Million:", correlation)
```

```python
print(correlation, ' is a very high negative relationship')
```

```python
# Scatter plot

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(df['Homes With'], df['Deaths per Mil'])
plt.xlabel('Homes With Smoke Detectors (%)')
plt.ylabel('Deaths per Million')
plt.title('Scatter Plot of Homes with Smoke Detectors vs. Deaths per Million')
plt.grid(True)
plt.show()
```
