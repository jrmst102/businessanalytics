```python
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd

# Note: We are switching the warnings off for purposes of clarity and cleanliness
import warnings
warnings.filterwarnings('error', category=DeprecationWarning)
```

```python
# Read data file from the Data folder

from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv('/content/drive/My Drive/Data/hyvee_stacked_r.csv')
df.head()
```

```python
# Examine which unique values we are dealing with
df['Product'].unique()
```

```python
# Prepare the data for Association Rules

df2 = pd.crosstab(df['Customer'], df['Product'])
df2
```

```python
# Hot Encode the data in case there is more than one transaction per customer, per item
def encode(item_freq):
    res = 0
    if item_freq > 0:
        res = 1
    return res

basket_input = df2.applymap(encode)
```

```python
# build the Apriori algorithm
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

frequent_itemsets = apriori(basket_input, min_support=0.001, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift")
rules.head()
```

```python
# Display the top rules
rules.sort_values(["support", "confidence","lift"],axis = 0, ascending = False).head(8)
```
