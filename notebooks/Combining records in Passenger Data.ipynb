```python
# Import Excel Files

import pandas as pd
sba_destination = pd.read_excel('sba494destination.xlsx')
sba_class = pd.read_excel('sba494class.xlsx')
```

```python
# Inspect Destination
sba_destination.head()
```

```python
# Inspect Class
sba_class.head()
```

```python
# Match rows by Confirmation

sba_merged = pd.merge(sba_class, sba_destination, on='Confirmation')
sba_merged.head()
```

```python
# Remove columns First_y and Last_y

sba_merged = sba_merged.drop(['First_y', 'Last_y'], axis=1)
sba_merged.head()
```

```python
#Save sba_merged as sba494.xlsx
sba_merged.to_excel("sba494.xlsx")
```
