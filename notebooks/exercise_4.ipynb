```python
# Open customers.csv
import pandas as pd
df = pd.read_csv("customers.csv")
df.head()
```

```python
# Visualize all columns name
for col in df.columns:
  print(col)
```

```python
# Display rows with missing values
df[df.isnull().any(axis=1)]
```

```python
# Drop rows with missing values
df = df.dropna()

# Display rows with missing values
df[df.isnull().any(axis=1)]
```

```python
# Trim spaces in all string columns

df_obj = df.select_dtypes(['object'])
df[df_obj.columns] = df_obj.apply(lambda x: x.str.strip())
df.head()
```

```python
# Capitalize letters in column Customer_Name
df['CUSTOMER_NAME'] = df['CUSTOMER_NAME'].str.capitalize()
df.head()
```

```python
# Convert Customer_Age from Float to Integer
df['CUSTOMER_AGE'] = df['CUSTOMER_AGE'].astype('int')
df.head()
```

```python
# Recode CUSTOMER_GENDER with F = 0 and M = 1
df['CUSTOMER_GENDER'] = df['CUSTOMER_GENDER'].replace({'F': 0, 'M': 1})
df.head()
```

```python
# Remove the , in annual purchases
df['ANNUAL_PURCHASES'] = df['ANNUAL_PURCHASES'].str.replace(',', '')
df['ANNUAL_PURCHASES'] = df['ANNUAL_PURCHASES'].str.replace('$', '')
df.head()
```

```python
# Convert ANNUAL_PURCHASES TO FLOAT

df['ANNUAL_PURCHASES'] = df['ANNUAL_PURCHASES'].astype('float')
df.head()
```

```python
# Create a new column where Adult = CUSTOMER_AGE < 65 and Senior = CUSTOMER_AGE >= 65
import numpy as np

df['CUSTOMER_TYPE'] = np.where(df['CUSTOMER_AGE']< 65, 'Adult', 'Senior')
df.head()
```

```python
# Show all the CUSTOMER_TYPE with 'Adult'

df_filtered = df[df['CUSTOMER_TYPE'] == 'Adult']
df_filtered.head()
```

```python
#Save the file as customers_clean.csv
df.to_csv('customers_clean.csv', index=False)
```
