```python
import pandas as pd
df = pd.read_excel('/content/sample_data/startingsalaries.xlsx')
print(df.head())
print(df['Monthly Starting Salary ($)'].map('{:.2f}'.format))
df['Monthly Starting Salary ($)'].describe()
q1 = df['Monthly Starting Salary ($)'].quantile(0.25)
q2 = df['Monthly Starting Salary ($)'].quantile(0.5)  # This is also the median
q3 = df['Monthly Starting Salary ($)'].quantile(0.75)

print("Q1:", q1)
print("Q2 (Median):", q2)
print("Q3:", q3)

# Calculate percentiles
for i in range(10, 100, 10):
  percentile = df['Monthly Starting Salary ($)'].quantile(i/100)
  print(f"{i}th percentile: {percentile}")
```

```python
# prompt: create a boxplot of monthly starting salaries

import matplotlib.pyplot as plt

plt.boxplot(df['Monthly Starting Salary ($)'], vert=False)
plt.title('Boxplot of Monthly Starting Salaries')
plt.ylabel('Monthly Starting Salary ($)')
plt.show()
```

```python
df2 = pd.read_csv('/content/electronics_r.csv')
print(df2.head())
```

```python
# prompt: calculate the covariance between no of commercials and sales volume

covariance = df2['No. of Commercials'].cov(df2['Sales Volume'])
print("Covariance between No of Commercials and Sales Volume:", covariance)
```

```python
# prompt: rename 'No. of Commercials' to 'Commercials'

df2 = df2.rename(columns={'No. of Commercials': 'Commercials', 'Sales Volume': 'Sales'})
print(df2.head())
```

```python
# prompt: plot commercials vs sales

# Assuming 'df2' is your DataFrame with 'Commercials' and 'Sales' columns
plt.figure(figsize=(10, 6))
plt.scatter(df2['Sales'], df2['Commercials'])
plt.ylabel('Number of Commercials')
plt.xlabel('Sales Volume')
plt.title('Scatter Plot of Commercials vs Sales')
plt.grid(True)  # Add a grid for better visualization
plt.show()
```
