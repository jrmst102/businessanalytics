```python
# Reading flights_r.csv

import pandas as pd
df = pd.read_csv('flights_r.csv')
df.head()
```

```python
# Calculate the mean of flights for the Atlanta column
print('Mean price of Atlanta tickets:', df['Atlanta'].mean())

# Calculate the mean of flights for the Salt Lake City column
print('Mean price of Salt Lake City tickets:', df['Salt_Lake_City'].mean())
```

```python
# Calculate the range, variance, and standard deviation for the two samples
print('Range of Atlanta tickets:', df['Atlanta'].max() - df['Atlanta'].min())
print('Variance of Atlanta tickets:', df['Atlanta'].var())
print('Standard deviation of Atlanta tickets:', df['Atlanta'].std())

print('Range of Salt Lake City tickets:', df['Salt_Lake_City'].max() - df['Salt_Lake_City'].min())
print('Variance of Salt Lake City tickets:', df['Salt_Lake_City'].var())
print('Standard deviation of Salt Lake City tickets:', df['Salt_Lake_City'].std())
```

```python
print('Atlanta tickets are less expensive and have less variability than Salt Lake City tickets')
```

```python
# EXTRA: Not included in the original problem
# Plot Atlanta Tickets and Salt Lake City tickets

import matplotlib.pyplot as plt

plt.hist(df['Atlanta'], alpha=0.5, label='Atlanta')
plt.hist(df['Salt_Lake_City'], alpha=0.5, label='Salt Lake City')
plt.legend(loc='upper right')
plt.title('Distribution of Flight Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```
