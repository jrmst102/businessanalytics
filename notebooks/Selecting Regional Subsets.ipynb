```python
# Read krsales_r.csv
import pandas as pd
df = pd.read_csv('krsales_r.csv')
df.head()
```

```python
# Select records where Region = Northeast
df_northeast = df[df['Region'] == 'Northeast']
df_northeast
```

```python
# Save Northeast records as pkrsalesne.csv
df_northeast.to_csv('pkrsalesne.csv', index=False)
```

```python
# Selecting records for East regions
print('Midwest')

df_midwest = df[df['Region'] == 'Midwest']
df_midwest
```

```python
# Selecting records for West regions
print('West')

df_west = df[df['Region'] == 'West']
df_west
```

```python
# Selecting records for South regions

print('South')

df_south = df[df['Region'] == 'South']
df_south
```

```python
# Saving West, Midwest, and South records
df_west.to_csv('pkrsalesw.csv', index=False)
df_midwest.to_csv('pkrsalesmw.csv', index=False)
df_south.to_csv('pkrsaless.csv', index=False)
```
