```python
import pandas as pd
employees = pd.read_csv("llemployees_r.csv")
employees.head()
```

```python
# Extract first two digits of EmpID and assign it to Day
employees['Day'] = employees['EmpID'].astype(str).str[:2]
employees.head()
```

```python
# Extract first two digits after the '-' and assign it to Month

employees['Month'] = employees['EmpID'].str.split('-').str[1].str[:2]
employees.head()
```

```python
# Extract four digits after the second '-'  and assign it to Year
employees['Year'] = employees['EmpID'].str.split('-').str[2]
employees.head()
```

```python
# Extract two digits after the third '-' and assign it to Department
employees['Department'] = employees['EmpID'].str.split('-').str[3].str[:2]
employees.head()
```

```python
# Extract last four digits of EmpId and assign it to Other
employees['Other'] = employees['EmpID'].str[-4:]
employees.head()
```

```python
# Optional
# Remove unnamed columns
employees = employees.loc[:, ~employees.columns.str.contains('^Unnamed')]
employees.head()
```

```python
# Optional
# Save file as 'employees_clean.csv'
employees.to_csv('employees_clean.csv')
```
