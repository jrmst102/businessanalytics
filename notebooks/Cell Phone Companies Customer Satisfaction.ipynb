```python
# prompt: read cellservice_r.csv file

import pandas as pd
df = pd.read_csv('cellservice_r.csv')
df.head()
```

```python
# Median Rating for T-Mobile
df['T-Mobile'].median()
```

```python
#Generate a five number summary for T-Mobile
df['T-Mobile'].describe()
```

```python
# Are there outliers in T-Mobile?
# We can calculate this by inspecting the numbers, by a box plot, or by using a statistical calculation.
# Here is the Box Plot
# Box Plot
import seaborn as sns
sns.boxplot(df['T-Mobile'])
```

In the plot above, there were no outliers.

```python
# Calculate the median ratings, five numbers summary, and box plot for AT&T, Sprint, and Verizon

#Median
print('Median for AT&T is', df['AT&T'].median())
print('Median for Sprint is', df['Sprint'].median())
print('Median for Verizon is', df['Verizon'].median())
```

```python
# 5-number summary
print('Five number summary for AT&T')
print(df['AT&T'].describe())
print('Five number summary for Sprint')
print(df['Sprint'].describe())
print('Five number summary for Verizon')
print(df['Verizon'].describe())
```

```python
# Boxplot for AT&T
sns.boxplot(df['AT&T'])
```

```python
# Boxplot for Sprint
sns.boxplot(df['Sprint'])
```

```python
# Boxplot for Verizon
sns.boxplot(df['Verizon'])
```

```python
# About the Box Plots
print('Verizon has the highest satisfation rating, followed by T-Mobile and then Sprint')
print('T-Mobile and Verizon have the lowest variability in ratings')
```
